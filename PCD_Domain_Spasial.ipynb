**A. Difference Image**

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img1 = cv2.imread('/content/kucing.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('/content/kucing.jpg', cv2.IMREAD_GRAYSCALE)

if img1 is None or img2 is None:
    print("Error: Citra tidak dapat dimuat.")
else:
    # Tambahkan noise ke citra kedua
    noise = np.random.normal(0, 25, img2.shape)
    img2_with_noise = cv2.add(img2, noise.astype(np.uint8))

    # Operasi pengurangan untuk mendapatkan citra difference
    diff_img = cv2.absdiff(img1, img2_with_noise)

    # Tampilkan citra difference
    plt.figure(figsize=(10, 10))
    plt.subplot(1, 3, 1), plt.imshow(img1, cmap='gray')
    plt.title('Citra 1'), plt.xticks([]), plt.yticks([])
    plt.subplot(1, 3, 2), plt.imshow(img2_with_noise, cmap='gray')
    plt.title('Citra 2 (with Noise)'), plt.xticks([]), plt.yticks([])
    plt.subplot(1, 3, 3), plt.imshow(diff_img, cmap='gray')
    plt.title('Difference Image'), plt.xticks([]), plt.yticks([])
    plt.show()
```

**B. Average Filter**

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load citra
img = cv2.imread('/content/kucing.jpg', cv2.IMREAD_GRAYSCALE)

# Periksa apakah citra berhasil dimuat
if img is None:
    print("Error: Citra tidak dapat dimuat.")
else:
    # Terapkan average filter dengan kernel 3x3
    kernel_size = 3
    kernel = np.ones((kernel_size, kernel_size), dtype=np.float32) / (kernel_size * kernel_size)
    filtered_img = cv2.filter2D(img, -1, kernel)

    # Tampilkan citra asli dan hasil filtering
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1), plt.imshow(img, cmap='gray')
    plt.title('Citra Asli'), plt.xticks([]), plt.yticks([])
    plt.subplot(1, 2, 2), plt.imshow(filtered_img, cmap='gray')
    plt.title('Hasil Average Filter'), plt.xticks([]), plt.yticks([])
    plt.show()
```

**C. Median Filter, Min Filter, dan Max Filter**

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load citra
img = cv2.imread('/content/kucing.jpg', cv2.IMREAD_GRAYSCALE)

if img is None:
    print("Error: Citra tidak dapat dimuat.")
else:
    # Terapkan Median Filter dengan kernel 3x3
    median_img = cv2.medianBlur(img, 3)

    # Terapkan Min Filter dengan kernel 3x3
    min_img = cv2.erode(img, np.ones((3,3),np.uint8))

    # Terapkan Max Filter dengan kernel 3x3
    max_img = cv2.dilate(img, np.ones((3,3),np.uint8))

    # Tampilkan citra asli dan hasil filtering
    plt.figure(figsize=(15, 10))

    plt.subplot(2, 2, 1), plt.imshow(img, cmap='gray')
    plt.title('Citra Asli'), plt.xticks([]), plt.yticks([])

    plt.subplot(2, 2, 2), plt.imshow(median_img, cmap='gray')
    plt.title('Median Filter'), plt.xticks([]), plt.yticks([])

    plt.subplot(2, 2, 3), plt.imshow(min_img, cmap='gray')
    plt.title('Min Filter'), plt.xticks([]), plt.yticks([])

    plt.subplot(2, 2, 4), plt.imshow(max_img, cmap='gray')
    plt.title('Max Filter'), plt.xticks([]), plt.yticks([])

    plt.tight_layout()
    plt.show()
```
